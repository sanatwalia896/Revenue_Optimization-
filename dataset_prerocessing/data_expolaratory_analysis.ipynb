```python
import pandas as pd
import numpy as np 
import seaborn as sns 
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
df=pd.read_csv("/Users/sanatwalia/Desktop/Assignments_applications/Revenue_Optimization-/dataset/mobile_game_iap_dataset_cleaned.csv")
```

```python
numeric_cols = df.select_dtypes(include='int64').columns

for col in numeric_cols:
    plt.figure(figsize=(8, 6))
    sns.histplot(df[col], kde=True, bins=30)
    plt.title(f'Distribution of {col}')
    plt.show()
```

```python
# Visualize categorical features
categorical_cols = ['User Segment', 'Purchase Type', 'Promo Applied', 'Item-Purchased']

for col in categorical_cols:
    plt.figure(figsize=(8, 6))
    sns.countplot(x=df[col])
    plt.title(f'Distribution of {col}')
    plt.show()
```

```python
relevant_corr=[]
for col in numeric_cols:
    if col not in categorical_cols:
        relevant_corr.append(col)
```

```python
# Correlation heatmap
df_corr=df[relevant_corr]
corr_matrix = df_corr.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```

```python
# Scatter plot between two features (example: Price vs Prior Purchases)
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Prior Purchases', y='Price (₹)', data=df)
plt.title('Price vs Prior Purchases')
plt.show()

# Pairplot to visualize relationships between multiple features
sns.pairplot(df[relevant_corr])
plt.show()
```

```python
# Average Price by User Segment
avg_price_by_segment = df.groupby('User Segment')['Price (₹)'].mean()
print(avg_price_by_segment)

# Average Price by Purchase Type
avg_price_by_purchase_type = df.groupby('Purchase Type')['Price (₹)'].mean()
print(avg_price_by_purchase_type)
# Average Price by Promo Applied            
avg_price_by_promo = df.groupby('Promo Applied')['Price (₹)'].mean()
print(avg_price_by_promo)       
# Average Price by Item Purchased
avg_price_by_item = df.groupby('Item-Purchased')['Price (₹)'].mean()
print(avg_price_by_item)

```

```python
"""After analyzing the correlation matrix and scatter plots, the following insights emerge:

There are minimal correlations between all variables in the dataset, with correlation coefficients near zero (-0.04 to 0.02).

The price variable shows a multimodal distribution with distinct pricing tiers (visible in the histogram).

Session times typically range between 50-75 minutes.

Level reached generally falls between 10-20.

Prior purchases appear as discrete values between 0-15.

Days installed before last purchase shows a relatively uniform distribution between 0-30 days.

The lack of strong correlations suggests that simple linear models may not capture the relationships effectively."""
```
